# Image to Sketch


```python
import cv2
# importing necessary libraries
import matplotlib.pyplot as plot
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/labrador.jpg")
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
invert_img = 225-gray_img
cv2_imshow(invert_img)
```

```python
img_smoothen = cv2.GaussianBlur(invert_img, (21,21), sigmaX = 0, sigmaY = 0)
cv2_imshow(img_smoothen)
```

```python
inverted_blurr = 255-img_smoothen
pencil_sketch = cv2.divide(gray_img, invert_img, inverted_blurr, scale = 256.0)
cv2_imshow(pencil_sketch)
```

```python
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```
